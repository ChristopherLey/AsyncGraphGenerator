```python

x = np.arange(0, 20)
y = np.cos(np.random.randn(x.shape[0]))
```

```python
tau = np.random.choice(y, size=4, replace=False)
t = []
for i in range(tau.shape[0]):
    t.append(np.where(tau[i] == y)[0])
t = np.concatenate(t)
```

```python
plt.figure(figsize=(40, 10))
plt.plot(x, y, lw=5, c='k')
plt.plot(t, tau, c='g', marker='o', ls='', ms=40, mfc='none', mew=10)
plt.show()
```
